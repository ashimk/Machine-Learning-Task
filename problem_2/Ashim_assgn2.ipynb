```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('problem_2_train.csv')
```

```python
df.head()
```

```python
df.describe()
```

```python
df.isnull().values.any()
```

```python
df.shape
```

```python
df.dtypes
```

```python
df.drop(columns=['hubpincode','memcreateddate','srno','mobilepurchasedate'],inplace=True)
```

```python
df.head()
```

```python
df.head()
```

```python
feature_cols = ['mobmake',
 'hubid',
 'pickuppartnerid',
 'deliverypartnerid',
 'insuranceCompanyId',
 'custpincodecategory',
 'claimtype',
 'custpincode',
 'servicecntrid',]
```

```python
X = df[feature_cols]
```

```python
y = df['tat_in_days']
```

```python
X.mobmake.value_counts()
```

# Label Encoding Categorical Data

```python
from sklearn.preprocessing import LabelEncoder
```

```python
labelEncoder = LabelEncoder()
```

```python
X['mobmake'] = labelEncoder.fit_transform(X['mobmake'])
```

```python
X.head()
```

```python
df.loc[df.claimtype != 'ADLD'].tat_in_days
```

```python
X.drop(columns=['claimtype'],inplace=True)
```

```python
X.head()
```

```python
X.custpincode.value_counts()
```

```python
X.custpincodecategory.value_counts()
```

```python
X.custpincodecategory = labelEncoder.fit_transform(X.custpincodecategory)
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(X,y,random_state=1,test_size=0.25)
```

# LinearRegression

```python
from sklearn.linear_model import LinearRegression
```

```python
linreg = LinearRegression()
```

```python
linreg.fit(x_train, y_train)
```

```python
y_pred = linreg.predict(x_test)
```

```python
from sklearn.metrics import mean_squared_error
```

```python
np.sqrt(mean_squared_error(y_pred,y_test))
```

```python
list(zip(feature_cols, linreg.coef_))
```

# RidgeRegression

```python
from sklearn.linear_model import Ridge
```

```python
ridge = Ridge()
```

```python
ridge.fit(x_train,y_train)
```

```python
y_pred = ridge.predict(x_test)
```

```python
np.sqrt(mean_squared_error(y_pred,y_test))
```

# GradientBoostRegression

```python
from sklearn.ensemble import GradientBoostingRegressor
```

```python
greg = GradientBoostingRegressor()
```

```python
greg.fit(x_train,y_train)
```

```python
y_pred = greg.predict(x_test)
```

```python
np.sqrt(mean_squared_error(y_pred,y_test))
```

# KNN

```python
from sklearn.neighbors import KNeighborsRegressor
```

```python
knn = KNeighborsRegressor()
```

```python
knn.fit(x_train,y_train)
```

```python
y_pred = knn.predict(x_test)
```

```python
np.sqrt(mean_squared_error(y_pred,y_test))
```
